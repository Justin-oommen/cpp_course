## Histograms

```python
res = np.random.randn(1000)*0.2 + 0.4
```

**Quick Histogram**

```python
plt.hist(res)
```

```python
a = plt.hist(res)
```

**change bins**

```python
plt.figure(figsize=(8,3))
plt.hist(res, bins=30,density=True)
plt.show()
```

The above plot is normalized to area 1 by using density=True.
This is called density plot.  
This is more popular because the density plots enables us to compare two histograms even though the comparing plots have different data set the density plot of the two of them is always normalized to area 1.

**Different style of bins(useful for plotting multiple histograms)**

```python
res2 = np.random.randn(1000)*0.2 + 0.4
```

```python
plt.figure(figsize=(8,3))
plt.hist(res, bins=30,density=True)
plt.hist(res2, bins=30,density=True)
plt.show()
```

Look here the two histograms cover each other and we have no idea what is going on with the blue one.  
  
We can use **histtype='step'**

```python
plt.figure(figsize=(8,3))
plt.hist(res, bins=30,density=True,histtype='step',label='1')
plt.hist(res2, bins=30,density=True,histtype='step',label='2')
plt.legend()
plt.show()

```

### StreamPlots

Really useful for vector fields

```python
w = 3
_ = np.linspace(-3,3,100)
x,y = np.meshgrid(_,_)
U = -1 - x**2 + y
v = 1 + x - y**2
speed = np.sqrt(U**2 + v**2)
```

```python
fig, axes = plt.subplots(2,2,figsize=(7,7))

ax = axes[0][0]
ax.streamplot(x,y,U,v)

ax = axes[0][1]
ax.streamplot(x,y,U,v,color=speed)

ax = axes[1][0]
lw = 5*speed / speed.max()
ax.streamplot(x,y,U,v,linewidth=lw)


ax = axes[1][1]
seedpoints = np.array([[1,0],[0,1]])
ax.streamplot(x,y,U,v,start_points=seedpoints)
```

### Animations

Animation require relating the time of our problem to the frame of the animation, then specifying the frames per second.  
  
- Below is a sin wave moving at 3 m/s. Translate this into an animation.

```python
def f(x,t):
    return np.sin(x-3*t) #wave equ = Asin(kx-wt) in positive x direction

x = np.linspace(0,10*np.pi,1000)
```

```python
plt.plot(x,f(x,0))
plt.plot(x,f(x,1))
plt.plot(x,f(x,2))
plt.plot(x,f(x,3))
```

**Create Animation**

```python
from matplotlib import animation
from matplotlib.animation import PillowWriter
```

Before making the animation we want to store empty line data
and time data.  
So we are going to diplay the time and and also plot the thing
but we want to make the objects but not want to initialize them.


```python
fig, ax = plt.subplots(1,1,figsize=(8,4))
ln1, = plt.plot([],[])
time_text = ax.text(0.65,0.95,'',fontsize=15,
                    transform=ax.transAxes,bbox=dict(facecolor='white',edgecolor='black'))

ax.set_xlim(0,10*np.pi)
ax.set_ylim(-1.5,1.5)

def animate(i):
    ln1.set_data(x,f(x,1/50*i))
    time_text.set_text(f"t={i/50:.2f}")
    
ani = animation.FuncAnimation(fig,animate,frames=250,interval=50)
ani.save('ani.gif',writer='pillow',fps=50,dpi=100)
```

Animation function can be used in other ways too. One can create a 3D rotating gifs of surfaces.
