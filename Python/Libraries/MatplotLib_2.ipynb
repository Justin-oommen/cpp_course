### Basic Graph

x and y are generally 1D array

```python
plt.plot([1,2,3],[2,4,6])
```

More Neatly

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y)

# title
plt.title('Our First Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
#x-label
plt.xlabel('X-axis',fontdict={'fontname':'cmmi10'})
#y-label
plt.ylabel('Y-axis',fontdict={'fontname':'cmmi10'})

plt.show()
```

To avoid the ticks(decimal values at the axes), we use `plt.xticks()` and `plt.yticks()`

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y,)

# title
plt.title('Our First Graph')
#x-label
plt.xlabel('X-axis')
#y-label
plt.ylabel('Y-axis')

#ticks
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])


plt.show()
```

- To add a legend we use `plt.legend()`
- use `color='our_desired_color'`for making the plot in our desired color.
- `linewidth=2`
- `marker='.'`
- `markersize=10`
- `linestyle='--'`
- `markeredgecolor='blue'`

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y,label='2x',color='red',linewidth=2,marker='.',markersize=10,markeredgecolor='blue',linestyle='--')

# title
plt.title('Our First Graph')
#x-label
plt.xlabel('X-axis')
#y-label
plt.ylabel('Y-axis')

#ticks
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

#legend
plt.legend()
plt.show()
```

A short hand notation for styling the plot  
`fmt='[color][marker][line]'`  
eg: `'r.--'`

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#plt.plot(x,y,label='2x',color='red',linewidth=2,marker='.',markersize=10,markeredgecolor='blue',linestyle='--')

plt.plot(x,y,'r^--',label='2x')

# title
plt.title('Our First Graph')
#x-label
plt.xlabel('X-axis')
#y-label
plt.ylabel('Y-axis')

#ticks
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

#legend
plt.legend()
plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#plt.plot(x,y,label='2x',color='red',linewidth=2,marker='.',markersize=10,markeredgecolor='blue',linestyle='--')

plt.plot(x,y,'r^--',label='2x')

#another plot
x2=np.arange(0,4.5,.5)
plt.plot(x2,x2**2,'g.-',label='x^2')


# title
plt.title('Our First Graph')
#x-label
plt.xlabel('X-axis')
#y-label
plt.ylabel('Y-axis')

#ticks
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

#legend
plt.legend()
plt.show()
```

If we want to make a projection of the graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#plt.plot(x,y,label='2x',color='red',linewidth=2,marker='.',markersize=10,markeredgecolor='blue',linestyle='--')

plt.plot(x,y,'r^--',label='2x')

#another plot
x2=np.arange(0,4.5,.5)
plt.plot(x2[:6],x2[:6]**2,'g.-',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r.--',label='x^2')


# title
plt.title('Our First Graph')
#x-label
plt.xlabel('X-axis')
#y-label
plt.ylabel('Y-axis')

#ticks
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

#legend
plt.legend()
plt.show()
```

To resize our graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#resize the graph(dpi specifies pixels per inch. When saving probably should use 300)
plt.figure(figsize=(5,3),dpi=100)

# Line 1

# Keyword argument Notation
#plt.plot(x,y,label='2x',color='red',linewidth=2,marker='.',markersize=10,markeredgecolor='blue',linestyle='--')


# Short hand Notation
plt.plot(x,y,'r^--',label='2x')

# Line 2

# another plot with an interval
x2=np.arange(0,4.5,.5)

# Plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2,'g.-',label='x^2')

# Plot remainder of the graph
plt.plot(x2[5:],x2[5:]**2,'g.--')


# title
plt.title('Our First Graph')
#x-label
plt.xlabel('X-axis')
#y-label
plt.ylabel('Y-axis')

# X and Y tickmarks(scale of our graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

# Legend
plt.legend()

# To save our graph (Saving in dpi=300 is in high resolution)
plt.savefig('mygraph.png',dpi=300)

# Show Plot
plt.show()
```

### Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

# Or we could do the above in a different way
patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4),dpi=100)

plt.show()
```

## Real World Examples

### Line Graph

```python
gas = pd.read_csv('gas_price.csv')
```

```python
gas
```

```python
print(gas.Year[::3])
```

```python
plt.figure(figsize=(8,4),dpi=200)

plt.title('Gas Prices over Time (in USD)')

plt.plot(gas.Year,gas.USA,'b.-')
plt.plot(gas.Year,gas.Canada,'m.-')
plt.plot(gas.Year,gas.Italy,'g.-')
plt.plot(gas.Year,gas['South Korea'],'r.-')

plt.xticks(gas.Year[::3])

labels=['USA','Canada','Italy','South Korea']
plt.legend(labels)

plt.show()
```

OR more efficiently to plot everything together

```python
for country in gas:
    if country != 'Year':
        print(country)
```

```python
plt.figure(figsize=(12,8),dpi=200)

plt.title('Gas Prices over Time (in USD)')

# Another way to plot multiple values

countries_to_look=['USA','Canada','Italy','South Korea']
for country in gas:
    if country in countries_to_look:
        plt.plot(gas.Year,gas[country],marker='.',label=country)
        
plt.xticks(gas.Year[::3])

#labels=['USA','Canada','Italy','South Korea']
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12,8),dpi=200)

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold','fontsize':15})

# Another way to plot all values

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker='.',label=country)
        
plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

To project it to 2011 we use `.tolist()+[2011]`

```python
plt.figure(figsize=(12,8),dpi=200)

plt.title('Gas Prices over Time (in USD)')

# Another way to plot multiple values

countries_to_look=['USA','Canada','Italy','South Korea']
for country in gas:
    if country in countries_to_look:
        plt.plot(gas.Year,gas[country],marker='.',label=country)
        
plt.xticks(gas.Year[::3].tolist()+[2011,2014])

plt.legend()

plt.savefig('gas_price.png',dpi=300)

plt.show()
```

#### Histograms

Google colorpicker to get color

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall,bins= bins,color='#97dff7')

plt.xticks(bins)

plt.ylabel('Number of players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.yticks()
plt.show()
```

#### Pie Charts

We would like to see percentage of people play with their preffered foot

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]

right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels = ['Left','Right']
colors = ['#f7e797','#bee5e6']

plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
# Reset the Weight which is now a strign to integer
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

# Plot style
plt.style.use('ggplot')

#fifa.Weight[0]


light = fifa.loc[fifa.Weight < 125.].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200.].count()[0]

light,light_medium,medium,medium_heavy,heavy

weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-150','150-175','175-200','Over 200']
explode=(.4,.2,0,0,.4)

plt.title('Weight Distribution of FIFA Players(in lbs)')

plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.8, explode=explode)

plt.show()
```

Compare (Relative Strengths) how good two teams - Box and whiskers chart (max,min,box around 50%)

```python
import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning) 
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']

real_madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']

revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']


labels = ['FC Barcelona', 'Real Madrid','NE Revolution']

#plt.boxplot([barcelona,real_madrid,revs],labels=labels)

#to style each individual, we save the boxplot in a variable
boxes = plt.boxplot([barcelona,real_madrid,revs],labels=labels,patch_artist=True,medianprops={'linewidth':2})

for box in boxes['boxes']:
    # set edge color and line width
    box.set(color='r',linewidth=2)
    
    # Change Fill color
    box.set(facecolor='#bee5e6')
    

plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
