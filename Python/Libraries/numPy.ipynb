Various mathematical function of numpy:  
- np.sum()
- np.prod()
- np.mean()
- np.std()
- np.var()
- np.min()
- np.max()
- np.argmin() gives index value of the minimum
- np.argmax() gives index value of the maximum

```python
print(np.sum(photo))
print(np.prod(photo))
print(np.mean(photo))
print(np.std(photo))
print(np.var(photo))
print(np.min(photo))
print(np.max(photo))
print(np.argmin(photo))
print(np.argmax(photo))
```
